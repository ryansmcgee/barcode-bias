# Process Diploid Bulk Fitness Assay (dBFA) data

```python
rawDataDir_dBFA2 = '/Users/ryan/Dropbox/Projects/Research/tikhonov/barcode-bias/barcode-bias/chen-case-study/data/chen-2023-elife-data/'
```

#### Load & organize raw data

##### Load the raw Chen et al. (2023) barcode counts data table

(rows: variants; cols: barcode counts for each experimental treatment/rep/timept and variant metadata)

```python
rawCountsTable_dBFA2 = pd.read_csv(rawDataDir_dBFA2+'dBFA2_counts_with_env_info.csv')
rawCountsTable_dBFA2
```

#### Extract the counts data (excluding metadata) from the variant and ancestor datasets

```python
metadata_columns = ['Full.BC', 'Diverse.BC', 'Environment.BC', 'Total.Counts', 'Subpool.Environment', 'Which.Subpools', 'Putative.Environment']
```

```python
rawCounts_dBFA2 = rawCountsTable_dBFA2.drop(columns=metadata_columns)
rawCounts_dBFA2
```

#### Extract the metadata for variants from the variants dataset

```python
variantsInfo_dBFA2 = rawCountsTable_dBFA2[metadata_columns].copy()  # this .copy() prevents variantsInfo_dBFA2 from being seen as a slice internally
variantsInfo_dBFA2 = variantsInfo_dBFA2.rename(columns={'Full.BC':'barcode_full', 'Diverse.BC':'barcode_diverse', 'Environment.BC':'barcode_environment', 'Total.Counts':'total_counts',
                                            'Subpool.Environment':'subpool_environment', 'Which.Subpools':'which_subpools', 'Putative.Environment':'putative_environment'})
variantsInfo_dBFA2
```

#### Parse treatment, rep, and timepoint information from sample labels (column names)

```python
sample_columns_raw = rawCounts_dBFA2.columns
sample_columns_raw.values
```

```python
samplesInfo_dBFA2 = []

sample_renames = {}

for sample in rawCounts_dBFA2.columns:
    sample_info = {}
    sample_info['chen_sample_label'] = sample
    sample_label_parts = sample.split('-')
    sample_info['assay_set'] = sample_label_parts[0]
    sample_info['treatment'] = '.'.join(sample_label_parts[1].split('_'))
    sample_info['rep']       = sample_label_parts[2]
    sample_info['timept']    = sample_label_parts[3][4:]
    sample_info['assay']     = sample_info['assay_set'] + ':' + sample_info['treatment'] + '-' + sample_info['rep']
    sample_info['sample']    = sample_info['assay'] + '-' + sample_info['timept']
    sample_info['total_raw_count'] = rawCounts_dBFA2[sample].sum()
    sample_renames[sample] = sample_info['sample']
    samplesInfo_dBFA2.append(sample_info)
    
samplesInfo_dBFA2 = pd.DataFrame(samplesInfo_dBFA2)

rawCounts_dBFA2 = rawCounts_dBFA2.rename(columns=sample_renames)
```

```python
samplesInfo_dBFA2['timept'] = pd.to_numeric(samplesInfo_dBFA2['timept'])
```

```python
# Sort dataframe for convenience:
samplesInfo_dBFA2.sort_values(by=['assay_set', 'treatment', 'rep', 'timept'], ascending=[True, True, True, True], ignore_index=True, inplace=True)
```

```python
# Reorder columns for cosmetic reasons
samplesInfo_dBFA2 = samplesInfo_dBFA2[['chen_sample_label', 'sample', 'assay', 'assay_set', 'treatment', 'rep', 'timept', 'total_raw_count']]
```

```python
samplesInfo_dBFA2
```

----------
#### Filter samples by total counts

```python
# Create a column in samplesInfo_dBFA2 to indicate which samples are excluded from analysis.
samplesInfo_dBFA2['exclude_from_analysis'] = False
```

```python
MIN_SAMPLE_COUNT_TOTAL = 1e5
```

```python
# Exclude treatments that have one or more samples with total count below threshold
good_coverage_samples = []
for treatment in samplesInfo_dBFA2['treatment'].unique():
    treatmentInfo = samplesInfo_dBFA2[samplesInfo_dBFA2['treatment'] == treatment]
    if(np.all((treatmentInfo['timept'] == 0) | (treatmentInfo['total_raw_count'] >= MIN_SAMPLE_COUNT_TOTAL))):
        good_coverage_samples.extend( treatmentInfo['sample'].values )
    else:
        samplesInfo_dBFA2.loc[samplesInfo_dBFA2['treatment'] == treatment, 'exclude_from_analysis'] = True
```

```python
rawCounts_dBFA2   = rawCounts_dBFA2[good_coverage_samples]
```

```python
# Drop excluded samples from the data set
samplesInfo_dBFA2 = samplesInfo_dBFA2[~samplesInfo_dBFA2['exclude_from_analysis']]
samplesInfo_dBFA2 = samplesInfo_dBFA2.drop(columns=['exclude_from_analysis'])
```

------

#### Get barcode sequence along with GC-content for each variant

```python
variants_GCratios = []
for idx, variant_info in variantsInfo_dBFA2.iterrows():
    variant_barcodeSeq = variant_info['barcode_full']
    variant_barcodeGCratio = (variant_barcodeSeq.count('G') + variant_barcodeSeq.count('C'))/len(variant_barcodeSeq)
    variants_GCratios.append(variant_barcodeGCratio)
#----------------
variantsInfo_dBFA2['barcode_GCratio']  = variants_GCratios
variantsInfo_dBFA2
```

#### Label control set variants ("putatively neutral barcodes")

```python
variantsInfo_dBFA2['neutral_group'] = ''
variantsInfo_dBFA2.loc[variantsInfo_dBFA2['subpool_environment'] == 'Ancestor_YPD_2N', 'neutral_group'] = 'Ancestor_YPD_2N'

variantsInfo_dBFA2['control_set'] = False
variantsInfo_dBFA2.loc[variantsInfo_dBFA2['subpool_environment'] == 'Ancestor_YPD_2N', 'control_set'] = True
```
